```python
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
df_hashTag = pd.read_csv('data/hashTag_L0.csv')

display(df_hashTag.head(5))
display(df_hashTag.info())
```

```python
df_hashTag.drop(columns=['R#'], inplace=True)

# --- Field Goals ---
df_hashTag['FGM'] = df_hashTag['FG%'].str.extract(r'\(([\d.]+)/')[0].astype(float)
df_hashTag['FGA'] = df_hashTag['FG%'].str.extract(r'/([\d.]+)\)')[0].astype(float)
df_hashTag['FG%'] = df_hashTag['FG%'].str.extract(r'([\d.]+)').astype(float)

# --- Free Throws ---
df_hashTag['FTM'] = df_hashTag['FT%'].str.extract(r'\(([\d.]+)/')[0].astype(float)
df_hashTag['FTA'] = df_hashTag['FT%'].str.extract(r'/([\d.]+)\)')[0].astype(float)
df_hashTag['FT%'] = df_hashTag['FT%'].str.extract(r'([\d.]+)').astype(float)
```

```python
df_hashTag = df_hashTag.rename(columns={'TREB': 'REB'})
df_hashTag = df_hashTag.rename(columns={'GP': 'G'})
```

```python
df_hashTag = df_hashTag[
    [
        'PLAYER', 'POS', 'TEAM', 'ADP',
        'G', 'MPG', 'PTS',
        'FGM', 'FGA', 'FG%', 
        'FTA', 'FTM', 'FT%', 
        '3PM', 'REB', 'AST', 'STL', 'BLK', 'TO', 'TOTAL'
    ]
]
```

```python
numeric_cols = [
    'ADP', 'G', 'MPG', 'PTS',
    'FGM', 'FGA', 'FG%', 
    'FTA', 'FTM', 'FT%', 
    '3PM', 'REB', 'AST', 'STL', 'BLK', 'TO', 'TOTAL'
]

# Convert safely â€” non-numeric entries become NaN
for col in numeric_cols:
    df_hashTag[col] = pd.to_numeric(df_hashTag[col], errors='coerce')
```

```python
df_hashTag[numeric_cols] = df_hashTag[numeric_cols].fillna(0)
```

```python
display(df_hashTag.head(5))
display(df_hashTag.info())
```

```python
# Convert to csv

df_hashTag.to_csv("../L1/data/hashTag_L1.csv")
```
