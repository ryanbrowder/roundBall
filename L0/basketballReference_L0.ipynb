```python
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
df_basketballReference = pd.read_csv('data/basketballReference_L0.csv')

display(df_basketballReference.head(5))
display(df_basketballReference.info())
```

```python
# Set the first row as header, then remove it from the data
df_basketballReference.columns = df_basketballReference.iloc[0]
df_basketballReference = df_basketballReference[1:].reset_index(drop=True)
```

```python
df_basketballReference = df_basketballReference.drop(columns=['Rk'])
df_basketballReference = df_basketballReference.drop(columns=['Type'])
df_basketballReference = df_basketballReference.drop(columns=['ORB'])
df_basketballReference = df_basketballReference.drop(columns=['PF'])
df_basketballReference = df_basketballReference.drop(columns=['-9999'])
```

```python
df_basketballReference = df_basketballReference.rename(columns={'Player': 'PLAYER'})
df_basketballReference = df_basketballReference.rename(columns={'FG': 'FGM'})
df_basketballReference = df_basketballReference.rename(columns={'FT': 'FTM'})
df_basketballReference = df_basketballReference.rename(columns={'3P': '3PM'})
df_basketballReference = df_basketballReference.rename(columns={'TRB': 'REB'})
df_basketballReference = df_basketballReference.rename(columns={'TOV': 'TO'})
```

```python
df_basketballReference = df_basketballReference[
    [
        'PLAYER',
        'PTS',
        'FGM', 'FGA', 'FG%', 
        'FTA', 'FTM', 'FT%', 
        '3PM', 'REB', 'AST', 'STL', 'BLK', 'TO'
    ]
]
```

```python
numeric_cols = [
        'PTS',
        'FGM', 'FGA', 'FG%', 
        'FTA', 'FTM', 'FT%', 
        '3PM', 'REB', 'AST', 'STL', 'BLK', 'TO'
]

# Convert safely — non-numeric entries become NaN
for col in numeric_cols:
    df_basketballReference[col] = pd.to_numeric(df_basketballReference[col], errors='coerce')
```

```python
display(df_basketballReference.head(5))
display(df_basketballReference.info())
```

```python
# Convert to csv

df_basketballReference.to_csv("../L1/data/basketballReference_L1.csv")
```
