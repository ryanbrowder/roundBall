```python
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
df_fanScout = pd.read_csv('data/fanScout_L0.csv')

display(df_fanScout.head(5))
display(df_fanScout.info())
```

```python
df_fanScout = df_fanScout.drop(columns=['RNK'])
df_fanScout = df_fanScout.drop(columns=['Value'])
df_fanScout = df_fanScout.drop(columns=['Status'])
df_fanScout = df_fanScout.drop(columns=['zPTS'])
df_fanScout = df_fanScout.drop(columns=['zTPM'])
df_fanScout = df_fanScout.drop(columns=['zAST'])
df_fanScout = df_fanScout.drop(columns=['zREB'])
df_fanScout = df_fanScout.drop(columns=['zSTL'])
df_fanScout = df_fanScout.drop(columns=['zBLK'])
df_fanScout = df_fanScout.drop(columns=['zTOV'])
df_fanScout = df_fanScout.drop(columns=['zFG%'])
df_fanScout = df_fanScout.drop(columns=['zFT%'])
```

```python
df_fanScout = df_fanScout.rename(columns={'Name': 'PLAYER'})
df_fanScout = df_fanScout.rename(columns={'M': 'MPG'})
df_fanScout = df_fanScout.rename(columns={'TPM': '3PM'})
df_fanScout = df_fanScout.rename(columns={'TOV': 'TO'})
```

```python
df_fanScout = df_fanScout[
    [
        'PLAYER', 'G',
        'MPG', 'PTS',
        'FGA', 'FG%', 
        'FTA', 'FT%', 
        '3PM', 'REB', 'AST', 'STL', 'BLK', 'TO'
    ]
]
```

```python
import numpy as np

numeric_cols = [
    'G','MPG', 'PTS',
    'FGA', 'FG%', 
    'FTA', 'FT%', 
    '3PM', 'REB', 'AST', 'STL', 'BLK', 'TO'
]

for col in numeric_cols:
    # Remove % and commas safely
    df_fanScout[col] = (
        df_fanScout[col]
        .astype(str)
        .str.replace('%', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.strip()
        .replace('', np.nan)
    )

    # Convert to numeric
    df_fanScout[col] = pd.to_numeric(df_fanScout[col], errors='coerce')

    # If column name includes '%', convert to decimal (e.g., 49 → 0.49)
    if '%' in col:
        df_fanScout[col] = df_fanScout[col] / 100
```

```python
display(df_fanScout.head(5))
display(df_fanScout.info())
```

```python
# Convert to csv

df_fanScout.to_csv("../L1/data/fanScout_L1.csv")
```
